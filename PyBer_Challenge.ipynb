### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os

# File to Load (Remember to check these)
city_data_to_load = os.path.join("Resources", "city_data.csv")
ride_data_to_load = os.path.join("Resources", "ride_data.csv")

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Challenge Deliverable 1. Generate a Ride-Sharing DataFrame by City Type

```python
#  1. Get the total rides for each city type
total_rides = pyber_data_df.groupby("type")["ride_id"].count()
total_rides
```

```python
# 2. Get the total drivers for each city type
total_drivers = city_data_df.groupby("type")["driver_count"].sum()
total_drivers
```

```python
#  3. Get the total amount of fares for each city type
total_fares = pyber_data_df.groupby(["type"])["fare"].sum()
total_fares
```

```python
#  4. Get the average fare per ride for each city type. 
average_fare_per_ride = pyber_data_df.groupby(["type"])["fare"].mean()
average_fare_per_ride
```

```python
# 5. Get the average fare per driver for each city type. 
average_fare_per_driver = total_fares/total_drivers
average_fare_per_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.concat([total_rides, total_drivers, total_fares, average_fare_per_ride, average_fare_per_driver], axis=1)
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
```

```python
#  8. Format the columns.
# Clean up the Index so that column names are unique
pyber_summary_df.columns = pd.Index(["a", "b", "c", "d", "e"])
# Use the rename() mathod
renaming = {"a": "Total Rides",
            "b": "Total Drivers",
           "c": "Total Fares",
           "d": "Average Fares per Ride",
           "e": "Average Fares per Driver"}
pyber_summary_df.rename(columns=renaming, inplace=True)
# See https://www.includehelp.com/python/format-a-number-with-commas-to-separate-thousands-in-pandas.aspx
# lambda = w00t

pyber_summary_df["Total Rides"] = pyber_summary_df["Total Rides"].apply(lambda datum: "{:,}".format(datum))
pyber_summary_df["Total Drivers"] = pyber_summary_df["Total Drivers"].apply(lambda datum: "{:,}".format(datum))
pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].apply(lambda datum: "$" + "{:,}".format(np.round(datum, decimals=2)))
pyber_summary_df["Average Fares per Ride"] = pyber_summary_df["Average Fares per Ride"].apply(lambda datum: "$" + "{:,.2f}".format(np.round(datum, decimals=2)))
pyber_summary_df["Average Fares per Driver"] = pyber_summary_df["Average Fares per Driver"].apply(lambda datum: "$" + "{:,.2f}".format(np.round(datum, decimals=2)))
pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# Print the merged DataFrame for reference.
pyber_data_df.head()
```

```python
# 1. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
fares_by_date = pyber_data_df.groupby(["type", "date"])["fare"].sum()
fares_by_date
```

```python
# 2. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
fares_by_date = fares_by_date.reset_index()
fares_by_date
```

```python
# 3. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
pivot_table_df = fares_by_date.pivot(index="date", columns="type", values="fare")
pivot_table_df
```

```python
# 4. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-28'.
pivot_table_jan_march_2019 = pivot_table_df.loc['2019-01-01':'2019-04-29', :]
pivot_table_jan_march_2019
```

```python
# 5. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
pivot_table_jan_march_2019.index = pd.to_datetime(pivot_table_jan_march_2019.index)
pivot_table_jan_march_2019
```

```python
# 6. Check that the datatype for the index is datetime using df.info()
pivot_table_jan_march_2019.info()
```

```python
# 7. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
pivot_table_by_week = pivot_table_jan_march_2019.resample("W").sum()
pivot_table_by_week
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

ax = pivot_table_by_week.plot(figsize=(20, 5))
ax.set_title("Fares")
ax.set_xlabel("")
ax.set_ylabel("Fare ($USD)")
```
